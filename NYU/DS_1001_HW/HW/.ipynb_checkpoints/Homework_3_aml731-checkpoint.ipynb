### Part 1 (5 Points)
Assume that $X$ and $Y$ are discrete random variables. The formulas for Mutual Information, Entropy and Conditional Entropy are given by:

$$\text{Mutual Information} = \sum_{y \in Y} \sum_{x \in X} p(x, y) \cdot log\frac{p(x, y)}{p(x)p(y)}$$

$$\text{Entropy} = H(Y) = -\sum_{y \in Y} p(y) \cdot log(p(y))$$

$$\text{Conditional Entropy} = H(Y \mid X) = \sum_{x \in X} p(x) \cdot H(Y \mid X = x)$$

Show mathematically that $\text{Mutual Information} = \text{Information Gain}$, where $\text{Information Gain} = H(Y) – H(Y \mid X)$. Give the derivation below (note, this can be done using Latek math notation, which renders nicely. See above. Feel free to do it by hand and submit an image of your proof).

__ANSWER__:

$$ \text{Information Gain} = H(Y) – H(Y \mid X)$$  
$$ =  -\sum_{y \in Y} p(y) \cdot log(p(y)) + \sum_{x \in X} p(x) \cdot H(Y \mid X = x) $$ 
$$= - \sum_{y \in Y} \sum_{x \in X} p(x,y) \cdot log(p(y)) + \sum_{x \in X} \sum_{y \in Y} p(x) p(y|x) \cdot log(p(y|x)) $$ 
$$= - \sum_{y \in Y} \sum_{x \in X} p(x,y) \cdot log(p(y)) + \sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot log(p(y|x)) $$ 
$$= \sum_{y \in Y} \sum_{x \in X} p(x,y) \cdot log\frac{p(y|x)}{p(y)}$$ 
$$= \sum_{y \in Y} \sum_{x \in X} p(x,y) \cdot log\frac{p(y|x)p(x)}{p(y)p(x)}$$ 
$$= \sum_{y \in Y} \sum_{x \in X} p(x,y) \cdot log\frac{p(x,y)}{p(x)p(y)} $$

### Part 3 - Finding a Good Decision Tree (10 Points)
The default options for your decision tree may not be optimal. We need to analyze whether tuning the parameters can improve the accuracy of the classifier.  For the following options `min_samples_split` and `min_samples_leaf`:

1\. Generate a list of 10 values of each for the parameters mim_samples_split and min_samples_leaf. 

````python
min_samples_split_values = [250,500,750, 1000, 2000,3000,4000,5000,7500,10000]
min_samples_leaf_values = [50 * (i+1) for i in range(10)]
````

__ANSWER__: One way to think about this is how many "profiles" of customers do we think there will be. The ranges we have considered allows roughly for ~100s of archetypes of customers, which feels like the right range (though we can probably be more precise with domain knowledge) since we want them to be general enough to apply to the test data but not overfit to small groups of custimers. 

We also can look at our original tree's depth and nodes above to get a sense of what we do NOT want, and ensure the depth and leaf sizes of our tree do not approach those values to avoid overfitting.

3\. For each combination of values in 3.1 (there should be 100), build a new classifier and check the classifier's accuracy on the test data. Plot the test set accuracy for these options. Use the values of `min_samples_split` as the x-axis and generate a new series (line) for each of `min_samples_leaf`.

__NOTE__: I updated my code to do several iterations of cross-validation for each set of parameters as I was noticing that there was a significant difference in the accuracy and optimal parameters for different splits of training/test data. The final accuracy score per set of parameters was an average of all the iterations. 

I kept the original code without these extra iterations in the comments in case that needed to be reviewed. 

````python
parameter_list = []

for split_value in min_samples_split_values:
    for leaf_value in min_samples_leaf_values:
        parameter_list.append((split_value,leaf_value))
        
parameter_df = pd.DataFrame(parameter_list)
parameter_df.columns = ['min_samples_split','min_samples_leaf']
````

````python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score
%matplotlib inline

def cell_data_dt_accuracy(X, Y, X_train, Y_train, X_test, Y_test, min_samples_split,min_samples_leaf):
    dt = DecisionTreeClassifier(criterion='entropy', min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf)
#     dt.fit(X_train, Y_train)
#     return accuracy_score(Y_test, dt.predict(X_test))
    cv_results = cross_validate(dt, X, Y, cv=5, scoring = 'accuracy', return_train_score=True)
    return np.mean(cv_results['test_score'])

# for i in range(len(parameter_df)):
#     min_samples_split = int(parameter_df.iloc[i]['min_samples_split'])
#     min_samples_leaf = int(parameter_df.iloc[i]['min_samples_leaf'])
#     test_accuracy = cell_data_dt_accuracy(X, Y, X_train, Y_train, X_test, Y_test, min_samples_split,min_samples_leaf)
#     parameter_df.loc[i,'test_accuracy'] = test_accuracy

randomized_df_list = []

for i in range(5):
    data_random = data.sample(frac=1)
    X = data_random.drop('churndep', 1)
    Y = data_random['churndep']
    randomized_df_list.append(data_random)
    
    
for i in range(len(parameter_df)):
    min_samples_split = int(parameter_df.iloc[i]['min_samples_split'])
    min_samples_leaf = int(parameter_df.iloc[i]['min_samples_leaf'])
    data_split_acc_list = []
    for df in randomized_df_list:
            X = df.drop('churndep', 1)
            Y = df['churndep']
            test_accuracy = cell_data_dt_accuracy(X, Y, X_train, Y_train, X_test, Y_test, min_samples_split,min_samples_leaf)
            data_split_acc_list.append(test_accuracy)
    parameter_df.loc[i,'test_accuracy'] = np.mean(data_split_acc_list)
````

````python
parameter_df[parameter_df['test_accuracy'] == parameter_df.test_accuracy.max()]
````

````python
plt.figure(figsize=(12,9))
for x in min_samples_leaf_values:
    plt.plot(parameter_df.min_samples_split[parameter_df['min_samples_leaf']==x], parameter_df.test_accuracy[parameter_df['min_samples_leaf']==x],label=x)

plt.legend(title='min_samples_leaf_values')
plt.xlabel('min_samples_split')
plt.ylabel('test_accuracy')
````
